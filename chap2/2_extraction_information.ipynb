```python
import pandas as pd
import numpy as np
import datetime as dt
```

# мӢңк°„ м •ліҙ м¶”м¶ң

мӢңк°„мқ„ н‘ңмӢңн•ҳлҠ” лҚ°мқҙн„° нғҖмһ…мқё datetime, Timestamp, datetime64мҷҖ мӢңк°„ лҚ°мқҙн„°мқҳ л°°м—ҙ мқёлҚұмҠӨмқё DatetimeIndex, PeriodIndex, мӢңк°„ к°„кІ© лҚ°мқҙн„°мқё timedelta, TimedeltaлҠ” лӢӨлҘё лҚ°мқҙн„° нғҖмһ…кіјлҠ” лӢ¬лҰ¬ лӮҙмһ¬лҗң м¶”к°Җм Ғмқё лҚ°мқҙн„° мҶҚм„ұл“Өмқҙ л§Һкі  мқҙм—җ лҢҖн•ң нҷңмҡ©лҸ„ л§Өмҡ° л№ҲлІҲн•ҳлӢӨ. к·ёл Үкё° л•Ңл¬ём—җ мӢңк°„ лҚ°мқҙн„°лҘј кө¬м„ұн•ҳлҠ” лӢӨм–‘н•ң лҚ°мқҙн„° мҶҚм„ұмқ„ мһҳ лӢӨлЈЁм–ҙм•ј н•ңлӢӨ. 

## Timestamp м •ліҙ м¶”м¶ң

pandasмқҳ TimestampлҠ” л…„, мӣ”, мқј, мӢң, л¶„, мҙҲ, л°ҖлҰ¬мҙҲ л“ұкіј к°ҷмқҙ лӮ м§ңмҷҖ мӢңк°„мқ„ кө¬м„ұн•ҳлҠ” м •ліҙлҘј к°Җм§Җкі  мһҲлӢӨ. н•ҳм§Җл§Ң мқҙ кё°мҙҲм Ғмқё м •ліҙмҷём—җлҸ„ л¶„кё° м •ліҙ, мҡ”мқј м •ліҙ, м—°мқҳ мӢңмһ‘мқјлЎңл¶Җн„° кІҪкіјмқјмҲҳ л“ұкіј к°ҷмқҖ лӮ м§ңмҷҖ мӢңк°„мқҙ к°Җм§ҖлҠ” нҠ№м„ұмқ„ мӮ¬мҡ©н•ҳлҠ” лӢӨм–‘н•ң м •ліҙл“Ө м¶”м¶ңн•  мҲҳлҸ„ мһҲлӢӨ. Timestampмқҳ м •ліҙлҠ” Timestamp лҚ°мқҙн„° нғҖмһ…мқҳ мҶҚм„ұ(attribute)мңјлЎңл¶Җн„° м–»лҠ” м •ліҙмҷҖ л©”мҸҳл“ң(method) н•ЁмҲҳлҘј мӮ¬мҡ©н•ҳм—¬ м–»лҠ” м •ліҙлЎң лӮҳлҲҢ мҲҳ мһҲлӢӨ. 

лӢӨмқҢмқҳ н‘ңлҠ” Timestamp лҚ°мқҙн„° нғҖмһ…мқҙ к°Җм§ҖлҠ” мЈјмҡ” мҶҚм„ұ м •ліҙмқҙлӢӨ.   

|мҶҚм„ұ|м„ӨлӘ…|
|:---:|:---:|
|year|м—°лҸ„ м •ліҙ|
|month|мӣ” м •ліҙ|
|day|мқј м •ліҙ|
|hour|мӢң м •ліҙ|
|minute|л¶„ м •ліҙ|
|second|мҙҲ м •ліҙ|
|microsecond|л§ҲмқҙнҒ¬лЎң мҙҲ м •ліҙ|
|dayofyear|м—°мқҳ кІҪкіј мқјмҲҳ м •ліҙ|
|week|м—°мқҳ кІҪкіј мЈјмҲҳ м •ліҙ|
|dayofweek|мЈјмқҳ мҡ”мқј м •ліҙ(мӣ” = 0, мқј = 6)|
|quarter|л¶„кё° м •ліҙ|
|is_leap_year|мңӨл…„мқём§Җм—җ лҢҖн•ң л…јлҰ¬к°’|

```python
pd.Timestamp('2023-02-04').dayofweek   ## Timestampмқҳ мҡ”мқј м •ліҙ м¶”м¶ң
```

```python
pd.Timestamp('2023-02-04').dayofyear   ##  Timestampмқҳ л…„ мӢңмһ‘мқјлЎңл¶Җн„° кІҪкіјмқј м •ліҙ м¶”м¶ң
```

```python
time_stamp = pd.Timestamp('2022-02-09')
print('{}, {} {}, {}'.format(time_stamp.day_name(), time_stamp.month_name(), time_stamp.day, time_stamp.year))
```

datetimeм—җм„ңлҠ” м ңкіөн•ҳлҠ” мҶҚм„ұмқҙ year, month, day, hour, minute, second, microsecond, tzinfoмҷҖ к°ҷмқҙ кё°ліём Ғмқё лӮ м§ңмҷҖ мӢңк°„ мҶҚм„ұмҷём—җлҠ” м ңкіөн•ҳм§Җ м•ҠлҠ”лӢӨ. л”°лқјм„ң лӮ м§ңмҷҖ мӢңк°„мңјлЎң нҢҢмғқлҗҳлҠ” мҶҚм„ұмқ„ м¶”м¶ңн•ҳкё° мң„н•ҙм„ңлҠ” datetimeмқ„ pandasмқҳ TimestampлЎң ліҖнҷҳн•ҳм—¬ м¶”м¶ңн•ҙм•ј н•ңлӢӨ. 

```python
pd.Timestamp(dt.date(2023, 2, 4)).week  ##  datetimeмқҳ м—°лҸ„ м •ліҙ м¶”м¶ң
```

лҳҗ numpyмқҳ datetime64 лҚ°мқҙн„° нғҖмһ…мқҖ мғқм„ұ мһҗмІҙк°Җ numpy к°қмІҙмқҙкё° л•Ңл¬ём—җ pandasмқҳ мҶҚм„ұл“Өмқ„ мӮ¬мҡ©н•ҳкё° мң„н•ҙм„ңлҠ” `astype()` л©”мҸҳл“ңм—җ 'object' л§Өк°ңліҖмҲҳлҘј м„Өм •н•ҳм—¬ лЁјм Җ python к°қмІҙмқё datetime к°қмІҙлЎң ліҖнҷҳн•ҙмЈјкі  мқҙлҘј лӢӨмӢң `Timestamp()`лҘј мӮ¬мҡ©н•ҳм—¬ pandas Timestamp к°қмІҙлЎң ліҖнҷҳн•ҙмӨҖлӢӨ. 

```python
pd.Timestamp(np.datetime64('2023-02-04').astype(object)).is_leap_year   ## datetime64мқҳ мқј м •ліҙ м¶”м¶ң
```

```python
np.datetime64('2023-02-04')
```

лӮ м§ңмҷҖ мӢңк°„ лҚ°мқҙн„° нғҖмһ…мқҙ python лӮҙл¶Җм—җ м ҖмһҘлҗ  л•ҢлҠ” 1970л…„ 1мӣ” 1мқјл¶Җн„°мқҳ кІҪкіјмқјмҲҳмҷҖ кІҪкіј лӮҳл…ёмҙҲ мҲҳлЎң м ҖмһҘлҗҳкё° л•Ңл¬ём—җ м •мҲҳлЎң ліҖнҷҳлҗҳм–ҙ м ҖмһҘлҗңлӢӨ. л”°лқјм„ң Timestamp лҚ°мқҙн„° нғҖмһ…мңјлЎң л°ҳнҷҳлҗҳлҠ” лӢӨм–‘н•ң мҶҚм„ұк°’мқҖ м •мҲҳнҳ• лҚ°мқҙн„°лӮҳ л…јлҰ¬к°’мңјлЎң л°ҳнҷҳлҗңлӢӨ. л§Ңм•Ҫ мӢңк°„ лҚ°мқҙн„° нғҖмһ…мқҙ к°Җм§ҖлҠ” м •мҲҳнҳ• кІ°кіјк°’мқҙ м•„лӢҢ мӢңк°„ лҚ°мқҙн„° мһҗмІҙм Ғмқё м„ұм§Ҳмқ„ ліҙмң н•ң м •ліҙ, мҰү мҡ”мқјмқҙлӮҳ мӣ” мқҙлҰ„ л“ұмқ„ м¶”м¶ңн•ҳкё° мң„н•ҙм„ңлҠ” мқҙлҘј мң„н•ҙ м ңкіөлҗҳлҠ” pandasмқҳ method н•ЁмҲҳлҘј мӮ¬мҡ©н•ңлӢӨ. лҳҗ мӢңк°„ лҚ°мқҙн„°лҘј мӮ¬мҡ©н•ҳкё° мң„н•ҙ н•„мҡ”н•ң м •ліҙлҘј м¶”м¶ңн•ҳкё° мң„н•ң лӢӨм–‘н•ң pandas method н•ЁмҲҳлҸ„ м ңкіөлҗҳлҠ”лҚ°, мЈјлЎң мӮ¬мҡ©лҗҳлҠ” method н•ЁмҲҳлҠ” лӢӨмқҢкіј к°ҷлӢӨ.

|combine()|date, time|лӮ м§ңмҷҖ м—°лҸ„лҘј лі‘н•©|

|strptime()|date_string, format|date_stringмқ„ formatм—җ л”°лқј н•ҙм„қн•ҳм—¬ datetime к°қмІҙ мғқм„ұ|
|to_numpy()||numpyмқҳ datetime64 к°қмІҙлЎң ліҖнҷҳ|

## DatetimeIndex м •ліҙ м¶”м¶ң

DatetimeIndexлҸ„ кё°ліём ҒмңјлЎң лӮ м§ңмҷҖ мӢңк°„ лҚ°мқҙн„° нғҖмһ…мқ„ кё°л°ҳмңјлЎң н•ҳкё° л•Ңл¬ём—җ лӮ м§ңмҷҖ мӢңк°„ нҠ№м„ұмқ„ н‘ңнҳ„н•ҳлҠ” мҶҚм„ұл“Өкіј л©”мҸҳл“ңлҘј м§Җмӣҗн•ңлӢӨ. лӢӨл§Ң DatetimeIndexлҠ” л°°м—ҙнҳ• мқёлҚұмҠӨмқҙкё° л•Ңл¬ём—җ мҶҚм„ұ м¶”м¶ң кІ°кіјлҸ„ м •мҲҳнҳ• лҳҗлҠ” л…јлҰ¬нҳ• лҚ°мқҙн„°мқҳ л°°м—ҙмқҙлӮҳ мқёлҚұмҠӨлҘј л°ҳнҷҳн•ңлӢӨ. лӢӨмқҢмқҖ DatetimeIndexм—җм„ң м§Җмӣҗн•ҳлҠ” мЈјмҡ” мҶҚм„ұмқҙлӢӨ. 

|мҶҚм„ұлӘ…|м„ӨлӘ…|
|---|---|
|year|м—°лҸ„ м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|month|мӣ” м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|day|мқј м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|hour|мӢң м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|minute|л¶„ м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|second|мҙҲ м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|dayofyear|м—°лҸ„ мӢңмһ‘мқјлЎңл¶Җн„° кІҪкіјмқј мқёлҚұмҠӨ м¶”м¶ң|
|dayofweek|мҡ”мқј м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|quarter|л¶„кё° м •ліҙ мқёлҚұмҠӨ м¶”м¶ң|
|is_month_start|н•ҙлӢ№ мӣ” мӢңмһ‘мқјмқём§Җм—җ лҢҖн•ң л…јлҰ¬к°’ мқёлҚұмҠӨ м¶”м¶ң|
|is_month_end|н•ҙлӢ№ мӣ” л§Ҳм§Җл§үмқјмқём§Җм—җ лҢҖн•ң л…јлҰ¬к°’ мқёлҚұмҠӨ м¶”м¶ң|
|is_leap_year|мңӨл…„ м—¬л¶Җ л…јлҰ¬к°’ л°°м—ҙ м¶”м¶ң|

```python
pd.to_datetime(['2023-02-04', '2023-02-05', '2023-02-06']).dayofweek  ## DatetimeIndexмқҳ м—°лҸ„ кІҪкіјмқјмҲҳ м¶”м¶ң
```

```python
pd.date_range("2023-02-04", periods = 30).is_month_start  ## DatetimeIndexмқҳ мЈј мҡ”м–ј м •ліҙ м¶”м¶ң
```

DatetimeIndexлҸ„ мҶҚм„ұкіј н•Ёк»ҳ м ңкіөлҗҳлҠ” л©”мҸҳл“ң н•ЁмҲҳл“Өмқҙ мһҲлӢӨ. мқҙ л©”мҸҳл“ң н•ЁмҲҳл“Өмқ„ мӮ¬мҡ©н•ҳм—¬ лӮ м§ңмҷҖ мӢңк°„мқҳ мҶҚм„ұмқ„ мӮ¬мҡ©н•ҙ мӮ°м¶ң к°ҖлҠҘн•ң лӢӨм–‘н•ң м •ліҙлҘј м¶”м¶ңн•ҙ лӮјмҲҳ мһҲлӢӨ. лӢӨмқҢмқҖ DatetimeIndexм—җм„ң м ңкіөлҗҳлҠ” мЈјмҡ” л©”мҸҳл“ң н•ЁмҲҳл“ӨмқҙлӢӨ.
